Categorizing Vote_Average column

We would cut the Vote_Average values and make 4 categories popular , average , below_avg , not_popular to describe it more using categorize_col() action provided above.

```python
import numpy as np
def catigorize_col(df, col, labels):

    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]
    edges = np.sort(np.unique(edges))

    df[col] = pd.cut(
        df[col],
        bins=edges,
        labels=labels[:len(edges)-1],
        include_lowest=True
    )
    return df



labels = ['not_popular', 'below_avg', 'average', 'popular']
catigorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
